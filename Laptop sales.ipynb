**IMPUTATION -**


The method of dealing with missing values by filling numerical columns with the median and categorical columns with the mode is referred to as Imputation. Specifically:

Median Imputation: For numerical columns, replacing missing values with the median is a robust method as it is not affected by outliers.



Mode Imputation: For categorical columns, replacing missing values with the mode ensures the most frequent category is used.
This combined approach is often referred to as Median-Mode Imputation or Simple Imputation in data preprocessing.

```python

# 3. Handle missing values - imputation
# - Fill numerical columns with median
# - Fill categorical columns with mode
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df[column] = df[column].fillna(df[column].median())

for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
def remove_outliers_iqr(df, column):
    """
    Removes outliers from a DataFrame column using the IQR method.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    column (str): The column name to check for outliers.

    Returns:
    pd.DataFrame: The DataFrame without outliers in the specified column.
    """
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound
    upper_bound = Q3 + 1.5 * IQR  # Upper bound


    # Filter the DataFrame to remove outliers
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# Apply the function to remove outliers in the 'Price column
data_no_outliers = remove_outliers_iqr(df, 'Price')

# Display the updated dataset's shape and a preview

print(f"Original dataset size: {df.shape[0]} rows")
print(f"Dataset size after outlier removal: {data_no_outliers.shape[0]} rows")
print(data_no_outliers.head())
```

.str.strip(): Removes any leading or trailing whitespace from each string in the Company column.


.str.title(): Converts the first character of each word in the Company column to uppercase and the remaining characters to lowercase.

This is typically done for data cleaning and standardization to ensure consistency in textual data. For instance, variations like "DELL", "dell", and " Dell " are all standardized to "Dell".




```python
#feature creation
# Define a function to categorize screen sizes
def categorize_screen_size(size):
    # Ensure 'size' is treated as a numeric value
    size = pd.to_numeric(size, errors='coerce')  # Convert to numeric, handle errors
    if size <= 14:
        return 'Small'
    elif size <= 16:
        return 'Medium'
    else:
        return 'Large'

# Apply the function to create a new column
df['Screen_Size_Category'] = df['Inches'].apply(categorize_screen_size)
df
```
